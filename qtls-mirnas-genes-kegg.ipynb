```python
import pandas as pd
import sqlite3
```

```python
connection = sqlite3.connect(":memory")
```

```python
qtls_mirnas = pd.read_csv('sql_qtl_mirnas_results.csv')
qtls_mirnas['miRNA_identifier'] = qtls_mirnas['miRNA_identifier'].str.replace('-UMD3.1.1','')
qtls_mirnas.to_sql('mirnas', connection, if_exists='replace')
```

```python
qtls_mirnas
```

```python
SQL_QUERY = '''
SELECT g.miRNA_identifier, g.gene_id, g.gene_description, m.miRNA_identifier,m.trait, m.QTL_id_UMD31,m.QTL_chr
FROM genes g JOIN mirnas m
ON g.miRNA_identifier = m.miRNA_identifier
AND g.QTL_id_UMD31 = m.QTL_id_UMD31

'''
final_df = pd.read_sql(SQL_QUERY, connection)
final_df
```
